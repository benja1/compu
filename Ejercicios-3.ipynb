#Ejercicio
Escribe un programa que agregue grafique en 2 columnas y 3 filas.

```python
x0=0
y0=0
vx0=40
vy0=70
x_out=[]
y_out=[]
g=9.81
T=20
t=0
ty=0
dt=0.001

while t<T:
    vy=vy0*ty-g*ty**2
    y=y0+vy
    if y<=0.0001:
        ty=0
        vy0=vy0-10
    x=x0+vx0*t
    t=t+dt
    ty=ty+dt
    x_out.append(x)
    y_out.append(y) 
  
        
plt.plot(x_out,y_out)
print x0, y0, vx0, vy0
```

Para gráficar una curva parametrica. 

```python
fig = plt.figure()#tamaño por default
ax = fig.gca(projection='3d')

theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)
ax.plot(x, y, z, label='parametric curve', linewidth = 3)
ax.legend() # escribe la etuiqueta de la grafica
```

```python
fig = plt.figure(figsize=(10,10))#tamaño por default
ax = fig.gca(projection='3d')

theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

#1 colored by value of `z`
ax.scatter(x, y, z, c = plt.cm.jet(z/max(z))) #puntos
```
